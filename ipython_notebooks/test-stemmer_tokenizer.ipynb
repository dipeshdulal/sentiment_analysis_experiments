```python
import re
```

```python
nepali_data = open('unchanged_articles.txt')
nepali_data = nepali_data.read().splitlines()
```

```python
def tokenizer(nepali_data):
    # sentence level
    ret_array = []
    for data in nepali_data:
        ret_array.append(re.split('(?<=[।?!]) +', data))
    return ret_array
```

```python
def stemmer(sentence):
    pattern=re.compile("(.*हरु)|(.*देखि)|(.*ेको)|(.*ेछु)|(.*ी)")
    ret_data = []
    sentence = sentence.split()
    for word in sentence:
        if pattern.match(word):
            ret_data.append(word)
    return ret_data
```

```python
tokenized_sentences = tokenizer(nepali_data)
```
